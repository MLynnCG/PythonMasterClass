## Ploting in Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,30,100)
y = np.sin(x)
```

## basic line plot

```python
plt.figure(figsize =(3,3))
plt.plot(x,y, marker = '*',color = 'red', linestyle ='-', label ='line plot', linewidth = 0.8)
plt.xlabel('x (m)', fontsize = 8)
plt.ylabel('f(x)', fontsize = 8)
plt.title('Basic Line plot', fontsize = 8)
plt.legend()
plt.grid()
plt.show()
```

## Subplot

```python
fig, axes = plt.subplots(2,2, figsize = (10,6))
axes[0,0].plot(x, y, 'r')
axes[0,0].set_title('Sin(x)')

axes[0,1].plot(x, np.tan(x), 'k')
axes[0,1].set_title('tan(x)')

axes[1,0].plot(x, np.cos(x), 'b')
axes[1,0].set_title('cos(x)')

axes[1,1].plot(x, -y, 'm')
axes[1,1].set_title('-sin(x)')

plt.tight_layout()
plt.show()
```

```python
x = np.random.rand(50) * 10  # Random x values
y = np.random.rand(50) * 20  # Random y values
sizes = np.random.rand(50) * 300  # Varying point sizes
colors = np.random.rand(50)  # Random colors

plt.figure()
plt.scatter(x, y, s=sizes, c=colors, cmap='jet', alpha=0.7, edgecolors='black')
plt.colorbar(label="Color Scale")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

```python
data = np.random.randn(1000)  # 1000 random values from a normal distribution

plt.figure()
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel("bin")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [3, 7, 1, 8]

plt.figure()
plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

```python
sizes = [20, 30, 25, 25]
labels = ["Apples", "Bananas", "Cherries", "Dates"]
colors = ['red', 'yellow', 'pink', 'brown']

plt.figure()
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Fruit Distribution")
plt.show()
```
