```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
 time = np.arange(0,100,1)
#time = range(0,100,5)
ages = np.random.rand(np.shape(time)[0])
ages_sqr = ages**2
```

```python
height = np.random.randn(2000)
```

```python
plt.figure()
hh = plt.hist(height,bins = 100)
plt.xlabel("height",fontsize = 18)
plt.ylabel("amount",fontsize = 18)
plt.title("Random Heights Test",fontsize = 18)
plt.tight_layout()
```

```python
plt.figure(figsize = (5,2))
plt.plot(time,ages,'b')
plt.plot(time,ages_sqr,'r')
plt.xlabel("time",fontsize = 18)
plt.ylabel("ages",fontsize = 18)
plt.title("Crazy Random Data Test",fontsize = 18)
plt.tight_layout()
```

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
z = y**2 + x**2
```

```python
plt.figure()
plt.scatter(x,y,s=5,c=z,cmap="jet")
plt.colorbar()
plt.xlabel("x",fontsize = 18)
plt.ylabel("y",fontsize = 18)
plt.title("Crazy Rando",fontsize = 18)
plt.tight_layout()
```
